```python
def oh_encode(col):
    b = pd.concat((train[col],test[col])).unique()
    # Sort and extract the indices.
    idx = sorted(b)
    # Initialise a matrix of zeros.
    aa = np.eye(len(idx))
    oh_dict = {}
    for i,k in enumerate(idx):
        oh_dict[k] = aa[i]
    return oh_dict
```

```python
def oh_encode_date(epoch):
    if epoch=="m":
        series = pd.DatetimeIndex(pd.concat((train["Date"],test["Date"])),dayfirst = True).month
    elif epoch=="y":
        series = pd.DatetimeIndex(pd.concat((train["Date"],test["Date"])),dayfirst = True).year
    # Sort and extract the indices.
    idx = sorted(series.unique())
    # Initialise a matrix of zeros.
    aa = np.eye(len(idx))
    oh_dict = {}
    for i,k in enumerate(idx):
        oh_dict[k] = aa[i]
    return oh_dict
```

```python
def get_datasets():
    league_encoding = oh_encode("LeagueId")
    month_encoding = oh_encode_date("m")
    year_encoding = oh_encode_date("y")
    home_encoding = oh_encode("HomeTeamId")
    away_encoding = oh_encode("AwayTeamId")
    train_list = []
    train_outputs = []
    test_list = []
    
    for index,row in train.iterrows():
        train_list.append(np.concatenate((league_encoding[row["LeagueId"]], month_encoding[row["Month"]], year_encoding[row["Year"]], home_encoding[row["HomeTeamId"]], away_encoding[row["AwayTeamId"]])))
        train_outputs.append(((row["Home_Corners"],row["Away_Corners"]),row["TOTAL_CORNERS"]))
        
    for index,row in test.iterrows():
        test_list.append(np.concatenate((league_encoding[row["LeagueId"]], month_encoding[row["Month"]], year_encoding[row["Year"]], home_encoding[row["HomeTeamId"]], away_encoding[row["AwayTeamId"]])))
            
    return train_list, train_outputs, test_list
```

```python
from scipy.stats import poisson
```

```python
def bet_exp(row):
    line = row["Line"]
    over = row["Over"]
    under = row["Under"]
    pred = row["Preds"]
    if line-int(line)==0: # i.e. asian line with possibility of push
        p_draw = poisson.pmf(line, pred)
        p_under = poisson.cdf(line-1, pred)
        p_over = 1-p_draw-p_under
    else: # no possibility of push
        p_under = poisson.cdf(int(line), pred)
        p_over = 1-p_under
        p_draw = 0
        
    exp_win = p_over*over + p_draw - 1
    exp_loss = p_under*under + p_draw - 1
    
    if exp_loss <0 and exp_win<0:
        outcome = np.nan
        stake = np.nan
    elif exp_loss>exp_win:
        outcome = "U"
        stake = (100*(1-p_over)//1)/10
    elif exp_win>exp_loss:
        outcome = "O"
        stake = (100*(1-p_under)//1)/10
    
    
    
    
    return [p_under, p_draw, p_over, exp_loss, exp_win, outcome, stake]
```
